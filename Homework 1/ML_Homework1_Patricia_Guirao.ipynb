# Question 2

```python
x_p2_p4 = (df['total_rooms']/df['population']).to_numpy().reshape(-1, 1)
y = df['median_house_value'].to_numpy().reshape(-1, 1)
model1 = LinearRegression().fit(x_p2_p4, y)
rSqr = model1.score(x_p2_p4, y)
print(rSqr)

yHat = model1.predict(x_p2_p4)
plt.plot(yHat,y,'o',markersize=.75)
plt.xlabel('Prediction from Model (Total Rooms/Population)') 
plt.ylabel('Median House Value')  
plt.title('R^2 = {:.3f}'.format(rSqr))
```

```python
x_p2_p5 = (df['total_rooms']/df['households']).to_numpy().reshape(-1, 1)
y = df['median_house_value'].to_numpy().reshape(-1, 1)
model1 = LinearRegression().fit(x_p2_p5, y)
rSqr = model1.score(x_p2_p5, y)
print(rSqr)

yHat = model1.predict(x_p2_p5)
plt.plot(yHat,y,'o',markersize=.75) 
plt.xlabel('Prediction from Model (Total Rooms/Households)') 
plt.ylabel('Median House Value')  
plt.title('R^2 = {:.3f}'.format(rSqr))
```
